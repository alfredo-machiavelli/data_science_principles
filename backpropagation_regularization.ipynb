```
def read_data(filename):
	'''
	Reads the input training data from filename and 
	Returns the matrices X : [N X D] and Y : [N X 1] where D is number of features and N is the number of data points
	'''
	dataframe = pd.read_csv(filename, keep_default_na=False, na_values='')
	data = [np.array(dataframe[col]) for col in dataframe]
	for i, d in enumerate(data):
		data[i].shape = (data[i].shape[0], 1)
	data = np.concatenate(data, axis = 1)
	X = data[:,:-1]
	Y = data[:,-1]
	Y.shape = (Y.shape[0],1)
	return X, Y

def one_hot_encode(X, labels):
	'''
	X = input [N X 1] matrix data 
	labels = list of all possible labels for current category
	Returns the matrix X : [N X len(labels)] in one hot encoded format
	'''
	X.shape = (X.shape[0], 1)
	newX = np.zeros((X.shape[0], len(labels)))
	label_encoding = {}
	for i, l in enumerate(labels):
		label_encoding[l] = i
	for i in range(X.shape[0]):
		newX[i, label_encoding[X[i,0]]] = 1
	return newX

def preprocess(X, Y):
	''' TASK 0
	X = input feature matrix [N X D] 
	Y = output values [N X 1]
	Convert data X, Y obtained from read_data() to a usable format by gradient descent function
	Return the processed X, Y that can be directly passed to grad_descent function
	NOTE: X has first column denote index of data point. Ignore that column 
	and add constant 1 instead (for bias part of feature set)
	'''
	inp_shape = X.shape
	preprocessed_X = np.ones((inp_shape[0],1))
	for dimension in range(1,len(X[0])):
		if(isinstance(X[0][dimension],int) or isinstance(X[0][dimension],float)):
			X_dimension = (X[:,dimension] - np.mean(X[:,dimension]))/np.std(X[:,dimension])
			X_dimension = np.reshape(X_dimension,[inp_shape[0],1])
			preprocessed_X = np.append(preprocessed_X,X_dimension,axis=1)
		elif(isinstance(X[0][dimension],str)):
			labels_dimension = list(set(list(X[:,dimension])))
			X_dimension = one_hot_encode(X[:,dimension],labels_dimension)
			preprocessed_X = np.append(preprocessed_X,X_dimension,axis=1)
	return preprocessed_X.astype(np.float),Y.astype(np.float)
```

```
X, y = read_data("house_prices_train.csv")
X, y = preprocess(X, y)
```

# Question 6 : Regularisation Guarantees (5 pts)

Consider the problem of linear regression with regularisation. We know that regularisation is meant to discourage large values of the weights - but is it guaranteed to do so? 

Concretely, consider the L2 regularisation introduced in class. Let the input data set be - $\{(\overrightarrow{x^{1}},y^{1}), (\overrightarrow{x^{2}},y^{2}), ...(\overrightarrow{x^{n}},y^{n})\}$, where $\overrightarrow{x^{r}} \in \mathbb{R}^{d}$ and $y^{r} \in \mathbb{R}$ (This basically means that each of the $x$'s are d-dimensional vectors and each of the $y$'s are scalars). Consider the following two different set of weights - 

$$ \overrightarrow{w^{1}} = argmin_{\overrightarrow{w} \in \mathbb{R}^{d}}\left(\sum_{r=1}^{n}(y^{r} - \overrightarrow{w}.\overrightarrow{x^{r}})^{2} \right) \;\;\;... (1)$$

$$ \overrightarrow{w^{2}} = argmin_{\overrightarrow{w} \in \mathbb{R}^{d}}\left(\sum_{r=1}^{n}(y^{r} - \overrightarrow{w}.\overrightarrow{x^{r}})^{2} \; \; + \; \lambda||\overrightarrow{w}||^{2} \right), \lambda > 0 \;\;\;... (2)$$

Here, $||\overrightarrow{x}||^{2} = x_{1}^{2} + x_{2}^{2} + x_{3}^{2} ... x_{d}^{2}$ represents the squared L2-norm of $\overrightarrow{x}$

The first equation represents the traditional unregularised linear regression with resulting weights $\overrightarrow{w^{1}}$, and the second objective represents the regularised linear regression with resulting weights $\overrightarrow{w^{2}}$.

Can you prove that $||\overrightarrow{w^{2}}||^{2} \leq ||\overrightarrow{w^{1}}||^{2}$? (This would imply that the weights learned after regularisation indeed have a smaller norm.)
 

## Hint :

On the face of it, it might appear intimidating, so let us work through some of the initial steps together.

Firstly, to simplify things, notice this expression, $ \sum_{r=1}^{n}(y^{r} - \overrightarrow{w}.\overrightarrow{x^{r}})^{2} $. It only has $\overrightarrow{w}$ as a variable, since all the $x's$ and $y's$ are just constant data points. So let's write it down as - $ f(\overrightarrow{w}) = \sum_{r=1}^{n}(y^{r} - \overrightarrow{w}.\overrightarrow{x^{r}})^{2} $

Using this notation, equations (1) and (2) can now be written as -

$$ \overrightarrow{w^{1}} = argmin_{\overrightarrow{w} \in \mathbb{R}^{d}}f(\overrightarrow{w}) \;\;\;... (3)$$

$$ \overrightarrow{w^{2}} = argmin_{\overrightarrow{w} \in \mathbb{R}^{d}}\left(f(\overrightarrow{w}) \; \; + \; \lambda||\overrightarrow{w}||^{2} \right), \lambda > 0 \;\;\;... (4)$$

Much better, right?

Ok, so now, we use the properties of $argmin$. 

Since $\overrightarrow{w^{1}}$ is the value of $\overrightarrow{w}$ at which $f(\overrightarrow{w})$ is smallest, we can write that 

$$ f(\overrightarrow{w^{1}}) \leq f(\overrightarrow{w^{2}}) \;\;\;... (5) $$

Similarly, Since $\overrightarrow{w^{2}}$ is the value of $\overrightarrow{w}$ at which $f(\overrightarrow{w}) + \; \lambda||\overrightarrow{w}||^{2}$ is smallest, we can write that 

$$ f(\overrightarrow{w^{2}}) +  \; \lambda||\overrightarrow{w^{2}}||^{2} \leq f(\overrightarrow{w^{1}}) + \; \lambda||\overrightarrow{w^{1}}||^{2} \;\;\;... (6) $$


Now your hint is, to use equations (5) and (6) to prove that prove that $||\overrightarrow{w^{2}}||^{2} \leq ||\overrightarrow{w^{1}}||^{2}$.
